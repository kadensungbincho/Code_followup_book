# Intro

- An operating system is software that manges a computer's hardware. It also provides a basis for application programs and acts as an intermediary between the computer user and the computer hardware.

- rough four parts: hardware, operating system, application programs, user

- In summary, for our purposes, the operating system includes the always-running kernel, middleware frameworks that ease application development and provide features, and system programs that aid in manaing the system while it is running.
